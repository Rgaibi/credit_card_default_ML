```python
df1 = data.copy()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df1, test_size=0.2, random_state=0)
```

```python
train_set.shape
```

```python
test_set.shape
```

```python
X_train = train_set.drop("default_payment_next_month", axis=1)
y_train = train_set["default_payment_next_month"]
X_test = test_set.drop("default_payment_next_month", axis=1)
y_test = test_set["default_payment_next_month"]
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
def decodage(df1) :
    for col in df1.select_dtypes("object").columns:
        df1[col] = LabelEncoder().fit_transform(df1[col])
    return df1
```

```python
def imputation(df1) :
    df1 = df1.drop("ID", axis=1)
    return df1
```

```python
def preprocessing(df1) :
    decodage(df1)
    imputation(df1)
    X = df1.drop("default_payment_next_month", axis=1)
    y = df1["default_payment_next_month"]
    return X, y
```

```python
X_train, y_train = preprocessing(train_set)
X_test, y_test = preprocessing(test_set)
```

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix, classification_report, f1_score, recall_score, precision_score
from sklearn.model_selection import learning_curve
```

```python
SELECTKBEST = SelectKBest(f_classif, k="all")
XGB = make_pipeline(SELECTKBEST, XGBClassifier(random_state=0))
RANDOMFOREST = make_pipeline(SELECTKBEST, RandomForestClassifier(random_state=0))
LOGISTICREGRESSION = make_pipeline(SELECTKBEST, StandardScaler(), LogisticRegression(random_state=0))
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
```

```python
dic_models = {"XGB" : XGB,
              "RANDOMFOREST" : RANDOMFOREST,
              "LOGISTICREGRESSION" : LOGISTICREGRESSION
             }
```

```python
def evaluation(model) :
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    train_size, train_score, validation_score = learning_curve(model, X_train, y_train, train_sizes=np.linspace(0.1,1,10), cv=cv, scoring="f1", random_state=0)
    plt.figure
    plt.plot(train_size, train_score.mean(axis=1), label="training score")
    plt.plot(train_size, validation_score.mean(axis=1), label="validation score")
    plt.legend()
    plt.show()
    plt.close()
```

```python
for key, model in dic_models.items() : 
    print(key)
    evaluation(model)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
XGB.get_params()
```

```python
hyper_params = {'selectkbest__k':[23,19],
                'xgbclassifier__learning_rate': [0.01],
                'xgbclassifier__n_estimators': [1000,2000],
                'xgbclassifier__max_depth': [3,5,7],
                'xgbclassifier__min_child_weight':[5,7,9],
                'xgbclassifier__subsample': [0.8,1],
                'xgbclassifier__colsample_bytree':[0.8,1],
                'xgbclassifier__gamma': [0,.01,1],
                'xgbclassifier__reg_lambda': [0,1,5],
                'xgbclassifier__scale_pos_weight': [3.5]
               }
```

```python
grid_r = RandomizedSearchCV(XGB, hyper_params, n_iter=20, scoring="recall", n_jobs=-1, cv=cv)
```

```python
grid_r.fit(X_train, y_train)
```

```python
grid_r.cv_results_['params']
```

```python
grid_r.best_params_
```

```python
evaluation(grid_r.best_estimator_)
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
y_prob = grid_r.best_estimator_.predict_proba(X_test)[:,1]
precision, recall, threshold = precision_recall_curve(y_test, y_prob)
plt.figure(figsize=(10,6))
plt.plot(threshold, precision[:-1], label="precision")
plt.plot(threshold, recall[:-1], label="recall")
plt.legend(loc="upper right")
plt.show()
plt.close()
```

```python
def model_final(model, X, threshold=0.4) :
    return model.predict_proba(X)[:,1]>0.4
```

```python
y_pred_final = model_final(grid_r.best_estimator_, X_test, threshold=0.4)
```

```python
print(confusion_matrix(y_test, y_pred_final))
print(classification_report(y_test, y_pred_final))
print(f'F1 Score: {f1_score(y_test, y_pred_final)}')
print(f'Recall: {recall_score(y_test, y_pred_final)}')
print(f'Precision: {precision_score(y_test, y_pred_final)}')
```
